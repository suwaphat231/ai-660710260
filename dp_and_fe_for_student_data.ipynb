```python
#เติมค่าอายุที่หายไปด้วยค่าเฉลี่ย
df['อายุ'] = df['อายุ'].fillna(df['อายุ'].mean())
print("หลังเติมค่า")
print(df)
```

```python
#เติมค่าเพศที่หายไปด้วยโหมด
df['เพศ'] = df['เพศ'].fillna(df['เพศ'].mode()[0])
print("หลังเติมค่า")
print(df)
```

```python
#แก้ไขค่าที่ผิดปกติ อายุเกิน 100
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]

print(f"\nขอบเขตล่าง - {lower_bound}")
print(f"ขอบเขตบน - {upper_bound}")
print(f"\nข้อมูลหลังกรองค่าผิดปกติ")
print(df_no_outliers)
print(f"\nค่าเฉลี่ยอายุ (หลังกรองค่าผิดปกติ) - {df_no_outliers['อายุ'].mean():.2f}")
```

```python
#แก้ไขค่าที่ผิดปกติ GPA
Q1 = df['GPA'].quantile(0.25)
Q3 = df['GPA'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_no_outliers = df[(df['GPA'] >= lower_bound) & (df['GPA'] <= upper_bound)]

print(f"\nขอบเขตล่าง - {lower_bound:.2f}")
print(f"ขอบเขตบน - {upper_bound:.2f}")
print(f"\nข้อมูลหลังกรองค่าผิดปกติ")
print(df_no_outliers)
print(f"\nค่าเฉลี่ย GPA (หลังกรองค่าผิดปกติ) - {df_no_outliers['GPA'].mean():.2f}")
```

```python
def normalize_gender(g):
    if pd.isna(g):
        return 'ไม่ระบุ'
    g = str(g).lower()
    if g in ['ชาย', 'ช', 'male', 'm']:
        return 'ชาย'
    elif g in ['หญิง', 'ญ', 'female', 'f']:
        return 'หญิง'
    else:
        return 'ไม่ระบุ'
```
